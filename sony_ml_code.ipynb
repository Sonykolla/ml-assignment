```python
# ============================================================
# TRANSFER LEARNING VS TRAINING FROM SCRATCH
# Topic:
# Transfer Learning vs Training a CNN from Scratch:
# What Small Datasets Can and Cannot Teach
#
# What this code does:
# 1. Builds a small image-classification dataset from CIFAR-10
# 2. Compares the same CNN architecture in two settings:
#       - random initialization (from scratch)
#       - pretrained weights (transfer learning)
# 3. Plots:
#       - training/validation accuracy
#       - training/validation loss
#       - confusion matrices
# 4. Produces a summary table for the report
#
# Notes:
# - Uses MobileNetV3-Small for faster runtime than larger CNNs
# - Uses a small subset of CIFAR-10 to simulate limited data
# - Resize is needed because pretrained ImageNet models expect larger images
# ============================================================

import os
import copy
import random
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms, models
from torchvision.models import MobileNet_V3_Small_Weights

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score
from sklearn.model_selection import train_test_split

# -----------------------------
# 1. Reproducibility
# -----------------------------
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# -----------------------------
# 2. Config
# -----------------------------
# Small subset sizes to simulate limited data
TRAIN_PER_CLASS = 150
VAL_PER_CLASS = 50
TEST_PER_CLASS = 100

# Keep only a few classes to make the task manageable and tutorial-friendly
# CIFAR-10 class ids:
# 0 airplane, 1 automobile, 2 bird, 3 cat, 4 deer, 5 dog, 6 frog, 7 horse, 8 ship, 9 truck
SELECTED_CLASSES = [0, 3, 5, 8]  # airplane, cat, dog, ship
CLASS_NAMES = ["airplane", "cat", "dog", "ship"]

BATCH_SIZE = 32
EPOCHS = 6           # Increase to 8-10 if you have GPU and more time
LEARNING_RATE = 1e-3
IMG_SIZE = 224       # Needed for pretrained ImageNet-style models

# -----------------------------
# 3. Transforms
# -----------------------------
# ImageNet-style normalization is used because the pretrained model expects it
imagenet_mean = [0.485, 0.456, 0.406]
imagenet_std  = [0.229, 0.224, 0.225]

train_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(imagenet_mean, imagenet_std)
])

eval_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(imagenet_mean, imagenet_std)
])

# -----------------------------
# 4. Load CIFAR-10
# -----------------------------
train_full = datasets.CIFAR10(root="./data", train=True, download=True, transform=train_transform)
train_full_eval = datasets.CIFAR10(root="./data", train=True, download=False, transform=eval_transform)
test_full = datasets.CIFAR10(root="./data", train=False, download=True, transform=eval_transform)

# -----------------------------
# 5. Build a small balanced subset
# -----------------------------
def collect_indices_by_class(targets, selected_classes):
    """
    Build a dictionary mapping each selected class to the list of indices
    belonging to that class.
    """
    class_to_indices = {c: [] for c in selected_classes}
    for idx, y in enumerate(targets):
        if y in selected_classes:
            class_to_indices[y].append(idx)
    return class_to_indices

train_class_indices = collect_indices_by_class(train_full.targets, SELECTED_CLASSES)
test_class_indices = collect_indices_by_class(test_full.targets, SELECTED_CLASSES)

# Choose fixed numbers per class for train/val/test
train_indices = []
val_indices = []
test_indices = []

for cls in SELECTED_CLASSES:
    cls_train_idx = train_class_indices[cls]
    cls_test_idx = test_class_indices[cls]

    # Split train into train/val
    chosen = cls_train_idx[:TRAIN_PER_CLASS + VAL_PER_CLASS]
    cls_train_split, cls_val_split = train_test_split(
        chosen,
        test_size=VAL_PER_CLASS,
        random_state=SEED,
        shuffle=True
    )

    train_indices.extend(cls_train_split[:TRAIN_PER_CLASS])
    val_indices.extend(cls_val_split[:VAL_PER_CLASS])
    test_indices.extend(cls_test_idx[:TEST_PER_CLASS])

# Use eval transform for val/test; use train transform for training
train_dataset = Subset(train_full, train_indices)
val_dataset   = Subset(train_full_eval, val_indices)
test_dataset  = Subset(test_full, test_indices)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

print(f"Train samples: {len(train_dataset)}")
print(f"Val samples:   {len(val_dataset)}")
print(f"Test samples:  {len(test_dataset)}")

# -----------------------------
# 6. Show a few training images
# -----------------------------
def denormalize(img_tensor, mean, std):
    """
    Convert normalized tensor back to displayable image.
    """
    img = img_tensor.clone().cpu().numpy().transpose(1, 2, 0)
    img = img * np.array(std) + np.array(mean)
    img = np.clip(img, 0, 1)
    return img

def show_batch(loader, class_names, n=8):
    """
    Display a small sample of training images.
    """
    images, labels = next(iter(loader))
    plt.figure(figsize=(14, 4))
    for i in range(min(n, len(images))):
        plt.subplot(2, 4, i + 1)
        plt.imshow(denormalize(images[i], imagenet_mean, imagenet_std))
        plt.title(class_names[SELECTED_CLASSES.index(labels[i].item())])
        plt.axis("off")
    plt.suptitle("Example images from the small dataset")
    plt.tight_layout()
    plt.show()

show_batch(train_loader, CLASS_NAMES)

# -----------------------------
# 7. Model builder
# -----------------------------
def build_model(pretrained=False, num_classes=4):
    """
    Build the same architecture in two modes:
    - pretrained=False: random initialization (from scratch)
    - pretrained=True: transfer learning with pretrained weights

    We replace the final classifier layer so both models learn the same target task.
    """
    if pretrained:
        weights = MobileNet_V3_Small_Weights.DEFAULT
        model = models.mobilenet_v3_small(weights=weights)
    else:
        model = models.mobilenet_v3_small(weights=None)

    # Replace the final classification layer
    in_features = model.classifier[-1].in_features
    model.classifier[-1] = nn.Linear(in_features, num_classes)

    return model

# -----------------------------
# 8. Training / evaluation helpers
# -----------------------------
def run_epoch(model, loader, criterion, optimizer=None):
    """
    Run one training or evaluation epoch.

    If optimizer is provided, the model trains.
    Otherwise, it runs in evaluation mode.
    """
    is_train = optimizer is not None
    model.train() if is_train else model.eval()

    total_loss = 0.0
    all_preds = []
    all_labels = []

    for images, labels in loader:
        # Remap original CIFAR labels to 0..num_classes-1
        labels = torch.tensor([SELECTED_CLASSES.index(l.item()) for l in labels], dtype=torch.long)

        images = images.to(device)
        labels = labels.to(device)

        if is_train:
            optimizer.zero_grad()

        with torch.set_grad_enabled(is_train):
            outputs = model(images)
            loss = criterion(outputs, labels)
            preds = outputs.argmax(dim=1)

            if is_train:
                loss.backward()
                optimizer.step()

        total_loss += loss.item() * images.size(0)
        all_preds.extend(preds.detach().cpu().numpy())
        all_labels.extend(labels.detach().cpu().numpy())

    epoch_loss = total_loss / len(loader.dataset)
    epoch_acc = accuracy_score(all_labels, all_preds)

    return epoch_loss, epoch_acc, np.array(all_labels), np.array(all_preds)

def train_model(model, train_loader, val_loader, epochs=6, lr=1e-3):
    """
    Train a model and keep history for report figures.
    """
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)

    history = {
        "train_loss": [],
        "train_acc": [],
        "val_loss": [],
        "val_acc": []
    }

    best_model_wts = copy.deepcopy(model.state_dict())
    best_val_acc = 0.0

    for epoch in range(epochs):
        train_loss, train_acc, _, _ = run_epoch(model, train_loader, criterion, optimizer=optimizer)
        val_loss, val_acc, _, _ = run_epoch(model, val_loader, criterion, optimizer=None)

        history["train_loss"].append(train_loss)
        history["train_acc"].append(train_acc)
        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)

        print(f"Epoch {epoch+1}/{epochs} | "
              f"train_loss={train_loss:.4f}, train_acc={train_acc:.4f} | "
              f"val_loss={val_loss:.4f}, val_acc={val_acc:.4f}")

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_model_wts = copy.deepcopy(model.state_dict())

    model.load_state_dict(best_model_wts)
    return model, history

def evaluate_on_test(model, test_loader):
    """
    Final test evaluation.
    """
    criterion = nn.CrossEntropyLoss()
    test_loss, test_acc, y_true, y_pred = run_epoch(model, test_loader, criterion, optimizer=None)
    return test_loss, test_acc, y_true, y_pred

# -----------------------------
# 9. Train from scratch
# -----------------------------
print("\n--- Training CNN from scratch ---")
scratch_model = build_model(pretrained=False, num_classes=len(SELECTED_CLASSES))
scratch_model, scratch_history = train_model(
    scratch_model,
    train_loader,
    val_loader,
    epochs=EPOCHS,
    lr=LEARNING_RATE
)

# -----------------------------
# 10. Train with transfer learning
# -----------------------------
print("\n--- Training with transfer learning ---")
transfer_model = build_model(pretrained=True, num_classes=len(SELECTED_CLASSES))
transfer_model, transfer_history = train_model(
    transfer_model,
    train_loader,
    val_loader,
    epochs=EPOCHS,
    lr=LEARNING_RATE
)

# -----------------------------
# 11. Plot accuracy curves
# -----------------------------
epochs_range = range(1, EPOCHS + 1)

plt.figure(figsize=(8, 5))
plt.plot(epochs_range, scratch_history["train_acc"], marker="o", label="Scratch train")
plt.plot(epochs_range, scratch_history["val_acc"], marker="o", linestyle="--", label="Scratch val")
plt.plot(epochs_range, transfer_history["train_acc"], marker="s", label="Transfer train")
plt.plot(epochs_range, transfer_history["val_acc"], marker="s", linestyle="--", label="Transfer val")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training and validation accuracy")
plt.legend()
plt.tight_layout()
plt.show()

# -----------------------------
# 12. Plot loss curves
# -----------------------------
plt.figure(figsize=(8, 5))
plt.plot(epochs_range, scratch_history["train_loss"], marker="o", label="Scratch train")
plt.plot(epochs_range, scratch_history["val_loss"], marker="o", linestyle="--", label="Scratch val")
plt.plot(epochs_range, transfer_history["train_loss"], marker="s", label="Transfer train")
plt.plot(epochs_range, transfer_history["val_loss"], marker="s", linestyle="--", label="Transfer val")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and validation loss")
plt.legend()
plt.tight_layout()
plt.show()

# -----------------------------
# 13. Final test evaluation
# -----------------------------
scratch_test_loss, scratch_test_acc, scratch_y_true, scratch_y_pred = evaluate_on_test(scratch_model, test_loader)
transfer_test_loss, transfer_test_acc, transfer_y_true, transfer_y_pred = evaluate_on_test(transfer_model, test_loader)

# -----------------------------
# 14. Confusion matrices
# -----------------------------
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

cm_scratch = confusion_matrix(scratch_y_true, scratch_y_pred)
disp1 = ConfusionMatrixDisplay(confusion_matrix=cm_scratch, display_labels=CLASS_NAMES)
disp1.plot(ax=axes[0], colorbar=False)
axes[0].set_title("Scratch CNN")

cm_transfer = confusion_matrix(transfer_y_true, transfer_y_pred)
disp2 = ConfusionMatrixDisplay(confusion_matrix=cm_transfer, display_labels=CLASS_NAMES)
disp2.plot(ax=axes[1], colorbar=False)
axes[1].set_title("Transfer Learning")

plt.tight_layout()
plt.show()

# -----------------------------
# 15. Summary table for report
# -----------------------------
scratch_best_train_acc = max(scratch_history["train_acc"])
scratch_best_val_acc = max(scratch_history["val_acc"])
transfer_best_train_acc = max(transfer_history["train_acc"])
transfer_best_val_acc = max(transfer_history["val_acc"])

summary_df = pd.DataFrame({
    "Model": ["CNN from scratch", "Transfer learning"],
    "Best Train Accuracy": [scratch_best_train_acc, transfer_best_train_acc],
    "Best Validation Accuracy": [scratch_best_val_acc, transfer_best_val_acc],
    "Test Accuracy": [scratch_test_acc, transfer_test_acc],
    "Final Train-Val Gap": [
        scratch_history["train_acc"][-1] - scratch_history["val_acc"][-1],
        transfer_history["train_acc"][-1] - transfer_history["val_acc"][-1]
    ]
})

print("\nSummary results:")
display(summary_df.round(4))

# -----------------------------
# 16. Small bar chart of final test accuracy
# -----------------------------
plt.figure(figsize=(6, 4))
plt.bar(summary_df["Model"], summary_df["Test Accuracy"])
plt.ylabel("Test accuracy")
plt.title("Final test accuracy comparison")
plt.tight_layout()
plt.show()

```
